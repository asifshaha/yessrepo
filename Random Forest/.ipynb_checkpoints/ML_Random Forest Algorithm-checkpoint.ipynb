# Random Forest

### Importing modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sb
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree
```

### Loading data

```python
data=pd.read_csv('data.csv')
print(data.head())
```

### Mapping Gender: Male:1, Female:0

```python
gen={'Male':1,'Female':0}
data['Gender']=data['Gender'].map(gen)
print(data.head())
```

### Assign Data to X and Y

```python
x=data[['Gender','Age','EstimatedSalary']]
y=data['Purchased']
```

### Split data for Train and test

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.3)
```

### RandomForestClassifier Model Create

```python
model=RandomForestClassifier(n_jobs=-1,random_state=42,n_estimators=50,min_samples_leaf=3,oob_score=True)
```

### Fit The Data

```python
model.fit(x_train,y_train)
```

### Predict the Y value

```python
y_predict=model.predict(x_test)
```

### Create DataFrame for compare y_test and y_predict

```python
result=pd.DataFrame({'Actual':y_test,'Prediction':y_predict})
print(result.head())
```

### Printing Confusion Matrix

```python
c_matrix=confusion_matrix(y_test,y_predict)
print('Confusion Matrix is :\n ',c_matrix)
```

###  Accuracy Score

```python
accuracy=accuracy_score(y_test,y_predict)*100
print('Accuracy Score is : ',accuracy)
```

###  Visualization

```python
pt.figure(figsize=(100,100))
plot_tree(model.estimators_[1],feature_names=x.columns,class_names=['1','0'],filled=True)
pt.show()
```
